```python
data = []

# Read each line of the JSONL file, parse JSON, and append to the list
with open('brands.json', 'r') as f:
    for line in f:
        json_obj = json.loads(line)
        data.append(json_obj)

# Convert the list of JSON objects to a DataFrame
df = pd.json_normalize(data)
```

# Receipts 

```python
# Receipts 
data1 = []

# Read each line of the JSONL file, parse JSON, and append to the list
with open('receipts.json', 'r') as f:
    for line in f:
        json_obj = json.loads(line)
        data1.append(json_obj)

# Convert the list of JSON objects to a DataFrame
df1 = pd.json_normalize(data1)
```

```python
df1.head()
```

```python

new_column_names1 = {
    '_id.$oid': '_id',
    'createDate.$date': 'createDate',
    'dateScanned.$date': 'dateScanned',
    'finishedDate.$date':'finishedDate',
    'modifyDate.$date':'modifyDate',
    'pointsAwardedDate.$date':'pointsAwardedDate',
    'purchaseDate.$date':'purchaseDate'
    
}

df1 = df1.rename(columns=new_column_names1)
```

```python
df1['createDate'] = pd.to_datetime(df1['createDate'], unit='ms')
df1['dateScanned'] = pd.to_datetime(df1['dateScanned'], unit='ms')
df1['finishedDate'] = pd.to_datetime(df1['finishedDate'], unit='ms')
df1['modifyDate'] = pd.to_datetime(df1['modifyDate'], unit='ms')
df1['pointsAwardedDate'] = pd.to_datetime(df1['pointsAwardedDate'], unit='ms')
df1['purchaseDate'] = pd.to_datetime(df1['purchaseDate'], unit='ms')
```
